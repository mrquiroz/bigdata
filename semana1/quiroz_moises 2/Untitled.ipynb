```python
import random
```

#### Ejercicio 1

```python
def create_random_row():
    # simulamos la columna edad
    age = random.randint(18, 90)
    # simulamos la columna ingreso
    income = random.randrange(10000, 1000000, step=1000)
    # simulamos la situación laboral
    employment_status = random.choice(['Unemployed', 'Employed'])
    # simulamos si es que tiene deuda o no
    debt_status = random.choice(['Debt', 'No Debt'])
    # simulamos si es que se cambió recientemente o no
    churn_status = random.choice(['Churn', 'No Churn'])
    return age, income, employment_status, debt_status, churn_status
```

se borra use_in_minutes debido a que no se define ni queda claro lo que significaria en caso de que se optara por generarla artificialmente

```python
# Se itera 1000 veces para no matar el computador
objeto = [list(create_random_row()) for i in range(60000)]
```

#### Ejercicio 3

```python
count_debts_looped = 0
for i in objeto:
    for j in i:
        if j == 'Debt':
            count_debts_looped += 1
```

¿Cuál es la complejidad algorítmica del código?

El orden de complejidad es de O(m*n) donde m corresponde a la cantidad de atributos en el objeto, ahora en el peor de los casos si m llega a ser del mismo largo de n, la complejidad pasaria a ser del orden de O($n^{2}$)

¿Cómo sería una implementación del código utilizando map y filter ?

```python
sum(map(lambda x:1,filter(lambda x :'Debt' in x, objeto)))
```

¿Cuál es la complejidad algorítmica del código?

La complejidad va del orden de 2m*n, donde m corresponde al numero de atribujos de cada uno de los i, asi en el peor de los casos la complejidad seria del orden de O($2n^{2}$)

¿Cómo sería una implementación del código utilizando map y filter ?

librerías externas.

para churn

```python
suma = sum(map(lambda x:x[0],churn))
promedio = suma/len(churn)
varianza = sum(map(lambda x: (x[0]-promedio)**2,churn))/len(churn)
minimo = min(map(lambda x:x[0],churn))
maximo = max(map(lambda x:x[0],churn))
```
